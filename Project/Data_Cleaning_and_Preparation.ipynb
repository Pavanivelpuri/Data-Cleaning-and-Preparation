**step 5: Missing value treatments**

```python
#choose the method of imputing missing value
#like mean,median,mode or KNNIputer

```

```python
# Strip spaces from column names
df.columns = df.columns.str.strip()

# Print columns to verify
print(df.columns)

# Fill missing values with the median for specified columns
for col in ["BMI", "Polio", "Income composition of resources"]:
    if col in df.columns:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        print(f"Column '{col}' not found in DataFrame")
```

```python
from sklearn.impute import KNNImputer
impute = KNNImputer()
```

```python
for i in df.select_dtypes(include='number').columns:
  df[i]=impute.fit_transform(df[[i]])
```

```python
df.isnull().sum()
```

**step 6:Outlier treatments**

```python
#decide whether to do outliers tratment or not, if do how?
```

```python
def wisker(col):
  q1,q3=np.percentile(col,[25,75])
  iqr=q3-q1
  uw=q3+1.5*iqr
  lw=q1-1.5*iqr
  return lw,uw
```

```python
wisker(df['GDP'])
```

```python
for i in ['GDP','Total expenditure','thinness  1-19 years','thinness 5-9 years']:
  lw,uw=wisker(df[i])
  df[i]=np.where(df[i]>uw,uw,np.where(df[i]<lw,lw,df[i]))
```

```python
for i in ['GDP','Total expenditure','thinness  1-19 years','thinness 5-9 years']:
  sns.boxplot(df[i])
  plt.show()
```
